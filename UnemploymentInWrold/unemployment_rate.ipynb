```python
# Filter data for each indicator
uem = info[info['IndicatorName'] == "Unemployment, youth total (% of total labor force ages 15-24) (national estimate)"]
salary = info[info['IndicatorName'] == "Wage and salaried workers, total (% of total employed)"]
services = info[info['IndicatorName'] == "Employment in services (% of total employment)"]
agriculture = info[info['IndicatorName'] == "Employment in agriculture (% of total employment)"]
industry = info[info['IndicatorName'] == "Employment in industry (% of total employment)"]
# Extract relevant columns for each indicator
uem = uem[['CountryCode','Year', 'Value']].rename(columns={'Value': 'Unemployment Value'})
salary = salary[['CountryCode', 'Year','Value']].rename(columns={'Value': 'Wage and salaried workers Value'})
services = services[['CountryCode', 'Year', 'Value']].rename(columns={'Value': 'Employment in services (% of total employment) Value'})
agriculture = agriculture[['CountryCode', 'Year', 'Value']].rename(columns={'Value': 'Employment in agriculture (% of total employment) Value'})
industry = industry[['CountryCode', 'Year', 'Value']].rename(columns={'Value': 'Employment in industry (% of total employment)'})
primaryeducation = primaryeducation[['CountryCode', 'Year', 'Value']].rename(columns={'Value': 'Unemployment with primary education Value'})
primaryeducation
```

```python
uem_2014 = uem[uem['Year'] == 2014].groupby('CountryCode')['Unemployment Value'].mean().reset_index()
salary_2014 = salary[salary['Year'] == 2014].groupby('CountryCode')['Wage and salaried workers Value'].mean().reset_index()
services_2014 = services[services['Year'] == 2014].groupby('CountryCode')['Employment in services (% of total employment) Value'].mean().reset_index()
agriculture_2014 = agriculture[agriculture['Year'] == 2014].groupby('CountryCode')['Employment in agriculture (% of total employment) Value'].mean().reset_index()
industry_2014 = industry[industry['Year'] == 2014].groupby('CountryCode')['Employment in industry (% of total employment)'].mean().reset_index()
primaryeducation_2014 = primaryeducation[primaryeducation['Year'] == 2014].groupby('CountryCode')['Unemployment with primary education Value'].mean().reset_index()
```

```python
merged_data = uem_2014.merge(salary_2014, on='CountryCode', how='outer')
merged_data = merged_data.merge(services_2014, on='CountryCode', how='outer')
merged_data = merged_data.merge(agriculture_2014, on='CountryCode', how='outer')
merged_data = merged_data.merge(industry_2014, on='CountryCode', how='outer')
merged_data = merged_data.merge(primaryeducation_2014, on='CountryCode', how='outer')
```

```python
merged_data
```

```python
merged_data = merged_data.rename(columns={
    'Unemployment Value': 'Unemployment Rate',
    'Wage and salaried workers Value': 'Wage and Salaried Workers',
    'Employment in services (% of total employment) Value': 'Employment in Services (%)',
    'Employment in agriculture (% of total employment) Value': 'Employment in Agriculture (%)',
    'Employment in industry (% of total employment)':'Employment in industry',
    'Unemployment with primary education Value':'Uemployment with primary education'
})
```

```python
y=merged_data['Unemployment Rate']
X=merged_data[['CountryCode','Wage and Salaried Workers', 'Employment in Services (%)', 'Employment in Agriculture (%)', 'Employment in industry']]
X=sm.add_constant(X)
X
```

There are some missing value, we have to take care of it first or there will have bugs.

```python
data = pd.concat([y, X], axis=1)
data = data.dropna()
data.sort_values(by='Unemployment Rate', ascending=True)['CountryCode'].head(10)
```

**Define y and X to the model**

```python
y = data['Unemployment Rate']
X = data[[ 'const', 'Wage and Salaried Workers', 'Employment in Services (%)', 'Employment in Agriculture (%)', 'Employment in industry']]
```

```python
model1 = sm.OLS(y, X)
results1 = model1.fit()
results1.summary()
```


**Wage and Servic modele**

```python
model2 = sm.OLS(data['Wage and Salaried Workers'], data['Employment in Services (%)'])
results2 = model2.fit()
results2.summary()
```
